```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
%matplotlib inline

def generate_Brownian():
    n_step = 1000
    x0 = float(0)
    w = np.ones(n_step)*x0
    for i in range(1,n_step):
        yi = np.random.normal()
        # Weiner process
        w[i] = w[i-1]+(yi/np.sqrt(n_step))
    return w
    
xaxis = np.arange(0, 2, float(1/500))
yaxis = generate_Brownian()
plt.plot(xaxis, yaxis)
plt.show()
```

```python
count_greater_than_pointSix = 0
count_B1_greater_than_zero_and_B2_greater_than_zero = 0

for i in range(1000):
    result = generate_Brownian()
    if any(num > 0.6  for num in result):
        count_greater_than_pointSix += 1
    if result[1] > 0 and result[2] > 0:
        count_B1_greater_than_zero_and_B2_greater_than_zero += 1
        
print(f"The calculated probability that Bt is greater than 0.6 is {round(count_greater_than_pointSix/1000, 3)}")
print(f"The calculated probability that B1 and B2 are both greater than 0 is {round(count_B1_greater_than_zero_and_B2_greater_than_zero/1000, 3)}")

# Actual probability of Bt > 0.6 before t = 2
value = 2*(1-norm.cdf(0.6))
print(f"The actual probability of Bt > 0.6 is {value}")
# Actual probability of B1 and B2 > 0
print(f"The actual probability that B1 and B2 > 0 is 0.375 (from class notes)")
```

From the above calculation, we can see the actual probability that Bt is > 0.6 before t = 2 is ~0.549. We are getting values close to this in our simulation.
From class notes, we know that P{B1 > 0, B2 > 0} = 3/8 = 0.375. We are seeing values very close to that in our simulations, or just slightly higher.
